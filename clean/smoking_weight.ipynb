## Question:
- Did quitting smoking cause people to gain weight?

## Scope
- The qsmk column indicated whether the subject quit smoking
- The wt82_71 column is the difference in weight between 1971 and 1982

## Motivation

## Mean difference

```python
ave_gain_quit = nhefs[nhefs.qsmk == 1].wt82_71.mean()
ave_gain_noquit = nhefs[nhefs.qsmk == 0].wt82_71.mean()

print("Average weight gain")
print("      quitters: {:>0.1f} kg".format(ave_gain_quit))
print("  non-quitters: {:>0.1f} kg".format(ave_gain_noquit))
```

```python
# Columns added for modeling
nhefs['constant'] = 1
nhefs['university'] = (nhefs.education == 5).astype('int')
nhefs['inactive'] = (nhefs.active == 2).astype('int')
nhefs['no_exercise'] = (nhefs.exercise == 2).astype('int')
```

## Confidence interval

```python
# Create a simple linear model to get a confidence interval on weight difference.
ols = sm.OLS(nhefs.wt82_71, nhefs[['constant', 'qsmk']])
res = ols.fit()
res.summary().tables[1]
```

```python
est = res.params.qsmk
conf_ints = res.conf_int(alpha=0.05, cols=None)
lo, hi = conf_ints[0]['qsmk'], conf_ints[1]['qsmk']

print('            estimate   95% C.I.')
print('difference   {:>6.2f}   ({:>0.1f}, {:>0.1f})'.format(est, lo, hi))
```

```python
summaries = OrderedDict((
    ('age', 'mean'),
    ('sex', lambda x: (100 * (x == 0)).mean()),
    ('race', lambda x: (100 * (x == 0)).mean()),
    ('university', lambda x: 100 * x.mean()),
    ('wt71', 'mean'),
    ('smokeintensity', 'mean'),
    ('smokeyrs', 'mean'),
    ('no_exercise', lambda x: 100 * x.mean()),
    ('inactive', lambda x: 100 * x.mean())
))

table = nhefs.groupby('qsmk').agg(summaries)
table.sort_index(ascending=False, inplace=True)
table = table.T

table.index = [
    'Age, years',
    'Men, %',
    'White, %',
    'University education, %',
    'Weight, kg',
    'Cigarettes/day',
    'Years smoking',
    'Little or no exercise, %',
    'Inactive daily life, %'
]

table.style.format("{:>0.1f}")
```

## PyMC Bayesian Model

```python
data = nhefs["wt82_71"]
data
```

```python
# No external regressors
with pm.Model() as model_simple:
    mu = pm.Uniform('mu', lower=-10, upper=10)
    sigma = pm.HalfNormal('sigma', sigma=5)
    Y = pm.Normal('Y', mu=mu, sigma=sigma, observed=data)
    idata_simple = pm.sample()
    idata_simple.extend(pm.sample_posterior_predictive(idata_simple))
```

```python
az.plot_trace(idata_simple);
```

```python
az.plot_ppc(idata_simple, num_pp_samples=100)
```

```python
dir(idata_simple.posterior)
```

```python
# No external regressors
with pm.Model() as model_smoke:
    alpha = pm.Normal("alpha", sigma= 4)
    beta = pm.Normal("beta", sigma = 4)
    sigma = pm.HalfNormal('sigma', sigma=5)
    mu = pm.Deterministic('mu', alpha + beta * nhefs.qsmk)
    y_pred = pm.Normal('Y', mu=mu, sigma=sigma, observed=data)
    idata_smoke = pm.sample()
    idata_smoke.extend(pm.sample_posterior_predictive(idata_smoke))
```

```python
# Beta shows the effect of quitting smoking on weight gain.
az.plot_trace(idata_smoke, var_names=["~mu"]);
```

```python
az.plot_ppc(idata_smoke, num_pp_samples=100);
```
