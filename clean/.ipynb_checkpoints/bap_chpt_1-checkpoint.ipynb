```python
def P(S, A):
    if set(A).issubset(set(S)):
        return len(A)/len(S)
    else:
        return 0
```

```python
def P(number_of_responses, yes_count):
    # Calculate the total number of combinations
    total_combinations = 2 ** number_of_responses
    
    # Calculate the number of combinations that have exactly 'yes_count' 'yes' responses
    specific_combinations = comb(number_of_responses, yes_count)
    
    # Calculate the probability of getting exactly 'yes_count' 'yes' responses
    probability = specific_combinations / total_combinations
    return probability

# Example usage:
number_of_responses = 3
yes_count = 2  # For example, calculating the probability of getting exactly 2 'yes' responses

probability = P(number_of_responses, yes_count)
print(f"Probability of getting exactly {yes_count} 'yes' responses out of {number_of_responses}: {probability}")
```

# Preliz

```python
x = np.linspace(0, 1, 100)

def a_preliz_model(a_mu, a_sigma, c_sigma=1):
    a = pz.Normal(a_mu, a_sigma).rvs()
    c = pz.Gamma(mu=2, sigma=c_sigma).rvs()
    b = pz.Normal(np.exp(a)*x, c).rvs()
    return b
```

```python
pz.predictive_explorer(a_preliz_model)
```

```python
plt.figure(figsize=(12, 9))

n_trials = [0, 1, 2, 3, 4, 8, 16, 32, 50, 150]
n_heads = [0, 1, 1, 1, 1, 4, 6, 9, 13, 48]
theta_real = 0.35

beta_params = [(0.5, 0.5), (20, 20), (1, 4)]
x = np.linspace(0, 1, 2000)

for idx, N in enumerate(n_trials):
    if idx == 0:
        plt.subplot(4, 3, 2)
        plt.xlabel('')
    else:
        plt.subplot(4, 3, idx+3)
        plt.xticks([])
    y = n_heads[idx]
    for (_prior, _prior) in beta_params:
        posterior = pz.Beta(_prior + y, _prior + N - y).pdf(x)
        plt.plot(x, posterior, lw=4)

    plt.plot(theta_real, 0, ms=9, marker='o', mec='w', mfc='k')
    plt.plot(0, 0, label=f'{N:4d} trials\n{y:4d} heads', alpha=0)
    plt.xlim(0, 1)
    plt.ylim(0, 12)
    plt.legend()
    plt.yticks([])

plt.savefig("../fig/ch1_q14_beta_binomial_posterior.png")
```
