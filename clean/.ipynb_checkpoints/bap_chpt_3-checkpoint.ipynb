```python
import numpy as np
import pymc as pm
import matplotlib.pyplot as plt
import pandas as pd
import preliz as pz
import arviz as az
```

```python
az.style.use("arviz-grayscale")
from cycler import cycler
default_cycler = cycler(color=["#000000", "#6a6a6a", "#bebebe", "#2a2eec"])
plt.rc('axes', prop_cycle=default_cycler)
plt.rc('figure', dpi=300)
```

```python
cs_data = pd.read_csv('../code/data/chemical_shifts_theo_exp.csv')
diff = cs_data.theo - cs_data.exp
cat_encode = pd.Categorical(cs_data['aa'])
idx = cat_encode.codes
coords = {'aa': cat_encode.categories}
```

```python
with pm.Model(coords=coords) as cs_nh:
    mu = pm.Normal('mu', mu=0, sigma=10, dims='aa')
    sigma = pm.HalfNormal('sigma', sigma=10, dims='aa')
    y = pm.Normal('y', mu=mu[idx], sigma=sigma[idx], observed=diff)
    idata_cs_nh = pm.sample()
```

```python
with pm.Model(coords=coords) as cs_h:
    # hyper_priors
    mu_mu = pm.Normal('mu_mu', mu=0, sigma=10)
    mu_sd = pm.HalfNormal('mu_sd', 10)
    #priors
    mu = pm.Normal('mu', mu=mu_mu, sigma=mu_sd, dims='aa')
    sigma = pm.HalfNormal('sigma', sigma=10, dims='aa')
    # likelihood
    y = pm.Normal('y', mu=mu[idx], sigma=sigma[idx], observed=diff)
    idata_cs_h = pm.sample()
```

```python
pm.model_to_graphviz(cs_nh)
```

```python
pm.model_to_graphviz(cs_h)
```

```python
axes = az.plot_forest([idata_cs_nh, idata_cs_h], model_names=['non_hierarchical', 'hierarchical'],
                      var_names='mu', combined=True, r_hat=False, ess=False, figsize=(10, 7),
                      colors='cycle')
y_lims = axes[0].get_ylim()
axes[0].vlines(idata_cs_h.posterior['mu_mu'].mean(), *y_lims, color="k", ls=":");
plt.savefig("../fig/csh_vs_csnh.png")
```

# Water Quality

## Generate synthetic data    

```python
# Total number of samples per group
N_samples = [30, 30, 30]
# Number of good-quality samples per group
G_samples = [3, 3, 18]
group_idx = np.repeat(np.arange(len(N_samples)), N_samples)
data = []
for i in range(0, len(N_samples)):
    data.extend(np.repeat([1, 0], [G_samples[i], N_samples[i] - G_samples[i]]))
```

```python
data[0:20]
```

```python
with pm.Model() as model_h:
    # hyperpriors
    mu = pm.Beta('mu', 1, 1)
    nu = pm.HalfNormal('nu', 10)
    # prior
    theta = pm.Beta('theta', mu = mu, nu = nu, shape = len(N_samples))
    # likelihood
    y = pm.Bernoulli('y', p = theta[group_idx], observed = data)

    idata_h = pm.sample()
```

```python
# G_samples = [18, 18, 18]
az.summary(idata_h)
```

# Hierarchical models

```python
football = pd.read_csv("../code/data/football_players.csv", dtype={'position':'category'})
football
```

```python
pos_idx = football.position.cat.codes.values
pos_codes = football.position.cat.categories
n_pos = pos_codes.size
n_players = football.index.size
```

```python
pos_codes
```

```python
pos_idx
```

```python
dist = pz.Beta()
# 95% of the mass is between 0 and 0.5
pz.maxent(dist, 0, 0.5, 0.95);
```

```python
dist_g = pz.Gamma()
# 95% of the mass is between 50 and 200
pz.maxent(dist_g, 50, 200, 0.95);
```

```python
coords = {"pos":pos_codes}
with pm.Model(coords=coords) as model_football:
    # Hyper parameters
    mu = pm.Beta('mu', 1.7, 5.8)
    nu = pm.Gamma('nu', mu = 125, sigma = 50)
    # Parameters for positions
    mu_p = pm.Beta('mu_p', mu = mu, nu = nu, dims = "pos")
    nu_p = pm.Gamma('nu_p', mu = 125, sigma = 50, dims = "pos")
    # Parameter for players
    theta = pm.Beta('theta', mu = mu_p[pos_idx], nu = nu_p[pos_idx])
    _ = pm.Binomial('gs', n = football.shots.values, p = theta, observed = football.goals.values)
    idata_football = pm.sample()
```

```python
# Global mean
idata_football.posterior.mean()
```

```python
# Forward player mean
idata_football.posterior.sel(pos="FW").mean()
```

```python
# Messi mean
idata_football.posterior.sel(theta_dim_0=1457).mean()
```

```python
_, ax = plt.subplots(3, 1, figsize=(12, 6), sharex=True)
az.plot_posterior(idata_football, var_names='mu', ax=ax[0])
ax[0].set_title(r"Global mean")
az.plot_posterior(idata_football.posterior.sel(pos="FW"), var_names='mu_p', ax=ax[1])
ax[1].set_title(r"Forward position mean")
az.plot_posterior(idata_football.posterior.sel(theta_dim_0=1457), var_names='theta', ax=ax[2])
ax[2].set_title(r"Messi mean")
# plt.savefig("../fig/beta_binomial_hierarchical_subjects_global_mus.png")
```
